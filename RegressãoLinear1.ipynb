```python
# Importação das bibliotecas, sklearn para criar o modelo de regressão e yellowbrick para visualização de residuais
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
# Aviso 
from yellowbrick.regressor import ResidualsPlot 
```

```python
 #Carregamento da base de dados e exclusão da primeira coluna
```

```python
cars= pd.read_csv('cars.csv')
```

```python
cars.head()
```

```python
cars.shape
```

```python
cars.drop(['Unnamed: 0'],axis=1, inplace=True)
cars
```

```python
# Definição das variáveis X e Y, X distância é a variável independente e Y velocidade é a variável dependente
```

```python
x=cars['dist'].values
y=cars['speed'].values
```

```python
# Cálculo da correlação entre X e Y
```

```python
correlacao = np.corrcoef(x,y)
correlacao
```

```python
#formato de matriz com uma coluna a mais
x = x.reshape(-1,1)
```

```python
# Criação do modelo e treinamento (.fit indica o inicio do treinamento
modelo = LinearRegression()
modelo.fit(x, y)
```

```python
# Visualização dos coeficientes
modelo.intercept_
```

```python
# inclinacao
modelo.coef_
```

```python
# Geração do gráfico com os pontos reais e as previsões
plt.scatter(x, y)
plt.title('Regressão Linear')
plt.xlabel('Distância')
plt.ylabel('Velocidade')
plt.plot(x, modelo.predict(x), color = 'red')
```

```python
# Previsão da "distância 30 pés" usando a fórmula manual
# interceptação * inclinação * valor de dist
#Qual velocidade se levou 30 pés pra parar?
modelo.intercept_ + modelo.coef_ * 30
```

```python
# Previsão utilizando função do sklearn
modelo.predict([[30]])
```

```python
# Gráfico para visualizar os residuais
visualizador = ResidualsPlot(modelo)
visualizador.fit(x,y)
visualizador.poof()
```
