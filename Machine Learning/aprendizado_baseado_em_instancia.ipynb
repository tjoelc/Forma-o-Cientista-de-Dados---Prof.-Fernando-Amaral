```python
# Importando as bibliotecas necessárias

import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn import datasets
from scipy import stats
```

```python
# vamos trazer a base Iris para teste
base = pd.read_csv('iris.csv')
base.head()
```

```python
base.describe()
```

```python
# carregando a base e visualização com o dataset
iris = datasets.load_iris()
stats.describe(iris.data)
```

```python
iris.data  # mesma coisa que o head so que aqui temos os array das instancias
```

```python
# Criaremos os previsores (Variavel independente x ) e class (Variavel dependente Y)
previsores = iris.data
classe = iris.target
```

```python
#Dividir a base para treinamento e teste 
x_treinamento,x_teste, y_treinamento,y_teste= train_test_split(previsores,classe,test_size=0.3,random_state= 0)
len(x_treinamento),len(x_teste)
```

```python
# vamos criar o modelo do treinamento agora
kncl= KNeighborsClassifier(n_neighbors=3) # criando o objeto e passado o numeros de vizinho para procurar a aparencia
kncl.fit(x_treinamento,y_treinamento) # passando os dados para treinamento
```

```python
# vamos as previsões
previsao = kncl.predict(x_teste) # passamos os dados para teste do modelo
previsao
```

```python
# vamos criar a matriz de confusão para avaliar 
confusao = confusion_matrix(y_teste,previsao)
confusao
```

```python
# vamos verificar a taxa de acerto desse algoritmo
tx_acerto = accuracy_score(y_teste,previsao)
tx_acerto
```
